## simple linear regression model(supervised ML)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
x=[1000,2000,3000,4000,6000,9000,8000,6600,7000,4400]
price=[2,3,4,5,6,7,8,33,9,22]
```

```python
sq_feet=np.array(x).reshape((-1,1))
```

```python
price=np.array(price)
```

```python
x_train,x_test,y_train,y_test=train_test_split(sq_feet,price,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
y_predict=model.predict(x_test)
```

```python
mse=mean_squared_error(y_predict,y_test)
mae=mean_absolute_error(y_predict,y_test)
r2=r2_score(y_predict,y_test)
```

```python
print("mean sq error",mse)
print("mean ab error",mae)
print("r2 score",r2)
```

```python
plt.scatter(sq_feet,price, label="Actual data",color="red")
plt.plot(sq_feet,model.predict(sq_feet),color="green",label="Regression line")
plt.xlabel("SQ feet")
plt.ylabel("price in lakhs")
plt.legend()
plt.show()
```

## Support Vector Machine(SVM)-Classification and Regression

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
iris=datasets.load_iris()
```

```python
x,y=iris.data,iris.target
```

```python
y
```

```python
pipe=pipeline([
    ('scalar',StandardScaler()),
    ('svm',SVC())
]
            )
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
grid={
    "svm__C":[0.1,1,10],
    "svm__kernel":['linear','rbf'],
    "svm__gamma":['scale','auto']

}
```

```python
grid_final=GridSearchCV(
    pipe,grid,cv=5
)
```

```python
grid_final.fit(x_train,y_train)
```

```python
y_predict=grid_final.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print("acc",accuracy_score(y_test,y_predict))
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```

## Classification Using Text

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
```

```python
data={
    "message":[
        "Hello,How  Are You",
        "you won 1crore",
        "Win money now",
        "You Got a Job In USA",
        "Have a Nice Day",
        "The Great Year Ahead"],
        "label":["ham","spam","spam","ham","ham","spam"]


}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df['label']=df['label'].map({
    "ham":0,
    "spam":1
})
```

```python
vect=CountVectorizer()
```

```python
x=df['message']
y=df['label']
vect_x=vect.fit_transform(x)
```

```python
print(vect_x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(vect_x,y,test_size=0.3,random_state=42)
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print("acc",accuracy_score(y_test,y_predict))
```

```python
new_data=["hi","can i take my pen"]
```

```python
x=vect.transform(new_data)
```

```python
f=model.predict(x)
```
