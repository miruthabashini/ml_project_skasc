```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BankChurners.csv")
```

```python
df["Churn"] = df["Attrition_Flag"].map({"Attrited Customer": 1, "Existing Customer": 0})
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="Gender", hue="Churn")
plt.title("Churn by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Churn")
plt.show()
```

```python
df["Churn"] = df["Attrition_Flag"].map({"Attrited Customer": 1, "Existing Customer": 0})
```

```python
churn_counts = df["Churn"].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    churn_counts,
    labels=["Stayed", "Churned"],
    autopct="%1.1f%%",
    colors=["lightgreen", "salmon"],
    explode=[0, 0.1],
    shadow=True
)
plt.title("Customer Churn Percentage", fontsize=16)
plt.show()
```

```python
import numpy as np
```

```python
df["Churn"] = np.where(df["Attrition_Flag"] == "Attrited Customer", 1, 0)
```

```python
avg_credit_limit = df.groupby("Months_on_book")["Credit_Limit"].mean()
```

```python

plt.figure(figsize=(8,5))
plt.plot(avg_credit_limit.index, avg_credit_limit.values, marker='o', color='green', linewidth=2)
plt.title("Average Credit Limit by Months on Book", fontsize=16)
plt.xlabel("Months on Book", fontsize=12)
plt.ylabel("Average Credit Limit", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
top10 = df.nlargest(10, "Credit_Limit").sort_values("Months_on_book")
```

```python
plt.scatter(
    top10["Months_on_book"],
    top10["Credit_Limit"],
    s=120,
    color="royalblue",
    edgecolor="black"
);
```
